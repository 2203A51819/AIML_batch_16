```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Convert categorical variables to numerical values
X_train_encoded = X_train.copy()
for column in X_train_encoded.columns:
    X_train_encoded[column] = label_encoder.fit_transform(X_train_encoded[column])

X_test_encoded = X_test.copy()
for column in X_test_encoded.columns:
    X_test_encoded[column] = label_encoder.transform(X_test_encoded[column])

nsimu = 21
accuracy = [0] * nsimu
ntree = [0] * nsimu
for i in range(1, nsimu):
    rfc = RandomForestClassifier(n_estimators=i * 5, min_samples_split=10, max_depth=None, criterion='gini')
    rfc.fit(X_train_encoded, y_train)
    rfc_pred = rfc.predict(X_test_encoded)
    cm = confusion_matrix(y_test, rfc_pred)
    accuracy[i] = (cm[0, 0] + cm[1, 1]) / cm.sum()
    ntree[i] = i * 5
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load your dataset for Chennai
df_chennai = pd.read_csv('/content/Air Quality Prediction Cleaned.csv')

# Check for missing values
print(df_chennai.isnull().sum())

# Drop rows with missing values
df_chennai.dropna(inplace=True)

# Assuming 'AQI_LVL' is the target variable and other columns are features
X_chennai = df_chennai[['PM2.5', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'VOC']]
y_chennai = df_chennai['AQI_LVL']

# Split the data into training and testing sets
X_train_chennai, X_test_chennai, y_train_chennai, y_test_chennai = train_test_split(X_chennai, y_chennai, test_size=0.2, random_state=42)

# Initialize the Random Forest Classifier
rf_classifier_chennai = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
rf_classifier_chennai.fit(X_train_chennai, y_train_chennai)

# Make predictions on the test set
y_pred_chennai = rf_classifier_chennai.predict(X_test_chennai)

# Calculate accuracy
accuracy_chennai = accuracy_score(y_test_chennai, y_pred_chennai)

# Calculate precision, recall, and F1-score for each class
precision_chennai = precision_score(y_test_chennai, y_pred_chennai, average=None)
recall_chennai = recall_score(y_test_chennai, y_pred_chennai, average=None)
f1_chennai = f1_score(y_test_chennai, y_pred_chennai, average=None)

# Calculate confusion matrix
conf_matrix_chennai = confusion_matrix(y_test_chennai, y_pred_chennai)

# Print the evaluation metrics for Chennai
print(f'Chennai - Accuracy: {accuracy_chennai}')
print(f'Chennai - Precision: {precision_chennai}')
print(f'Chennai - Recall: {recall_chennai}')
print(f'Chennai - F1-score: {f1_chennai}')
print(f'Chennai - Confusion Matrix:\n{conf_matrix_chennai}')
```
